```python
#Which month has the most disasters

import pandas as pd
import matplotlib.pyplot as plt

# Supondo que seu DataFrame Ã© chamado df
disasters_columns['start_month'] = disasters_columns['start_month'].astype(int)

# Contagem de ocorrÃªncias de desastres por mÃªs
monthly_counts = disasters_columns['start_month'].value_counts().sort_index()


month_names = [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]

# CriaÃ§Ã£o do grÃ¡fico de barras
plt.figure(figsize=(10, 6))
plt.bar(month_names, monthly_counts, color='skyblue')
plt.title('Number of Disasters per Month')
plt.xlabel('Month')
plt.ylabel('Number of Disasters')
plt.show()
```

```python
floods_df = disasters_columns[disasters_columns['disaster_type'] == 'Flood']

# Group by country and count the occurrences of floods
country_flood_counts = floods_df['country'].value_counts()

# Get the country with the maximum number of floods
country_with_most_floods = country_flood_counts.idxmax()

# Print the result
print(f"The country with the most floods is: {country_with_most_floods}")
```

```python
floods_df = disasters_columns[(disasters_columns['disaster_type'] == 'Flood') & (disasters_columns['year'] >= 2016) & (disasters_columns['year'] <= 2021)]

# Group by country and count the occurrences of floods
country_flood_counts = floods_df.groupby('country').size().reset_index(name='count')

# Sort the data by count in descending order
country_flood_counts = country_flood_counts.sort_values(by='count', ascending=False)

# Select the top 5 countries
top5_countries = country_flood_counts.head(5)

# Create the bar plot for the top 5 countries
plt.figure(figsize=(8, 8))
sns.barplot(x='country', y='count', data=top5_countries, palette='viridis')  # Swapped x and y here
plt.title('Top 5 Countries with the Most Floods (2016-2021)')
plt.ylabel('Number of Floods')  # Changed to match Y axis
plt.xlabel('Country')  # Changed to match X axis
plt.show()
```

```python
# Filter rows for floods and the years 2016 to 2021
floods_df = disasters_columns[(disasters_columns['disaster_type'] == 'Flood') & (disasters_columns['year'] >= 2016) & (disasters_columns['year'] <= 2021)]

# Group by country and count the occurrences of floods
country_flood_counts = floods_df.groupby('country').size().reset_index(name='count')

# Sort the data by count in ascending order to get the least affected countries
country_flood_counts = country_flood_counts.sort_values(by='count')

# Select the bottom 5 countries
bottom5_countries = country_flood_counts.head(5)

# Create the bar plot for the bottom 5 countries
plt.figure(figsize=(12, 6))
sns.barplot(x='count', y='country', data=bottom5_countries, palette='viridis')
plt.title('Top 5 Countries with the Fewest Floods (2016-2021)')
plt.xlabel('Number of Floods')
plt.ylabel('Country')
plt.show()
```

```python
# Filter for floods between 2016 and 2021
floods_df = disasters_columns[(disasters_columns['disaster_type'] == 'Flood') & 
                              (disasters_columns['year'] >= 2016) & 
                              (disasters_columns['year'] <= 2021)]

# Group by country and sum the deaths
country_flood_deaths = floods_df.groupby('country')['total_deaths'].sum().reset_index()

# Sort the data by deaths in descending order
country_flood_deaths = country_flood_deaths.sort_values(by='total_deaths', ascending=False)

# Select the top 5 countries
top5_countries = country_flood_deaths.head(5)

# Create the bar plot for the top 5 countries
plt.figure(figsize=(8, 8))
sns.barplot(x='country', y='total_deaths', data=top5_countries, palette='viridis')  # Swapped x and y here
plt.title('Top 5 Countries with Most Deaths Due to Floods (2016-2021)')
plt.ylabel('Number of Deaths')  # Changed to match Y axis
plt.xlabel('Country')  # Changed to match X axis
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# To make sure month columns are integer type
disasters_columns['start_month'] = disasters_columns['start_month'].astype(int)
disasters_columns['end_month'] = disasters_columns['end_month'].astype(int)

# Create a new column combining year and month to facilitate temporal analysis
disasters_columns['year_month'] = disasters_columns['year'].astype(str) + '-' + disasters_columns['start_month'].astype(str).str.zfill(2)

# Count of occurrences of each type of disaster per month
monthly_disaster_counts = disasters_columns.groupby(['year_month', 'disaster_type']).size().unstack(fill_value=0)

# Identify the most common type of disaster each month
most_common_disaster_by_month = monthly_disaster_counts.idxmax(axis=1)

# Create a bar chart to show the most common disaster type by month
plt.figure(figsize=(12, 8))
sns.set_palette("husl")
monthly_disaster_counts.plot(kind='bar', stacked=True)
plt.title('Most Common Type of Disaster by Month Over the Years')
plt.xlabel('Year and Month')
plt.ylabel('Number of Occurrences')
plt.legend(title='Disaster Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Merge Geospatial Data with Disaster Data
merged_data = world.merge(disasters_columns, how='left', left_on='name', right_on='country')
merged_data
```

```python
#Map Distribution of Different Types of Disasters

import matplotlib.pyplot as plt


fig, ax = plt.subplots(1, 1, figsize=(15, 10))
merged_data.plot(column='disaster_type', legend=True, ax=ax, legend_kwds={'bbox_to_anchor': (1, 1)})
plt.title('Distribution of Different Types of Disasters')
plt.show()
```

```python
#Interactive Maps with Plotly

import plotly.express as px

fig = px.choropleth(merged_data, geojson=merged_data.geometry, locations=merged_data.index,
                    color='disaster_type', projection='natural earth')
fig.update_geos(showcoastlines=True, coastlinecolor="black", showland=True, landcolor="lightgray")
fig.update_layout(title='Distribution of Different Types of Disasters')
fig.show()
```

```python
#To identify the most occurred disaster type in each year
#Grouping by Year and Disaster Type
grouped_data = disasters_columns.groupby(['year', 'disaster_type']).size().reset_index(name='occurrences')
```

```python
#Identifying the Most Occurred Disaster in Each Year
idx = grouped_data.groupby(['year'])['occurrences'].transform(max) == grouped_data['occurrences']
most_occurred_per_year = grouped_data[idx]

#most_occurred_per_year contains the most occurred disaster type in each year along with the number of occurrences of that disaster type.
```

```python
#Visualization or Additional Analysis

# Example visualization using matplotlib
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
for year, group in most_occurred_per_year.groupby('year'):
    plt.bar(year, group['occurrences'], label=f'{year}: {group["disaster_type"].values[0]}')

plt.xlabel('Year')
plt.ylabel('Number of Occurrences')
plt.title('Most Occurred Disaster in Each Year')
plt.legend()
plt.show()
```

```python
import geopandas as gpd
import plotly.express as px

# Group by Year, Country, and Disaster Type and count the number of occurences 
grouped_data = disasters_columns.groupby(['year', 'country', 'disaster_type']).size().reset_index(name='occurrences')

# Identify the Most Occurred Disaster in Each Country and Year
idx = grouped_data.groupby(['year', 'country'])['occurrences'].transform(max) == grouped_data['occurrences']
most_occurred_per_country_year = grouped_data[idx]

# Load Geospatial Data
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create Maps for Each Year
#O cÃ³digo, entÃ£o, itera pelos anos Ãºnicos presentes no DataFrame
#most_occurred_per_country_year e cria mapas choropleth para cada ano.
for year in most_occurred_per_country_year['year'].unique():
    filtered_data = most_occurred_per_country_year[most_occurred_per_country_year['year'] == year]
    
    # Merge Geospatial Data with Disaster Data
    merged_data = world.merge(filtered_data, how='left', left_on='name', right_on='country')
    
    # Visualization with Plotly Express
    fig = px.choropleth(merged_data, geojson=merged_data.geometry, locations=merged_data.index,
                        color='disaster_type', projection='natural earth',
                        title=f'Most Occurred Disaster Type in Each Country in {year}')
    fig.update_geos(showcoastlines=True, coastlinecolor="black", showland=True, landcolor="lightgray")
    #fig.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# For Storms
df_storm = disasters_columns[disasters_columns['disaster_type'] == 'Storm'].groupby('country').size().reset_index(name='total')

# Check the available data
if not df_storm.empty:
    g_storm = df_storm[df_storm['total'] > 15].sort_values('total', ascending=False) \
        .assign(country=lambda x: pd.Categorical(x['country'], categories=x['country'][::-1], ordered=True)) \
        .plot(kind='barh', x='country', y='total', color='skyblue', legend=False)
    plt.title("Total number of Storms per country\nfrom 1901 to 2021")
    plt.xlabel("Total")
    plt.ylabel("Country")
    plt.show()
else:
    print("No data available for Storms.")

# For Floods
df_flood = disasters_columns[disasters_columns['disaster_type'] == 'Flood'].groupby('country').size().reset_index(name='total')

# Check the available data
if not df_flood.empty:
    g_flood = df_flood[df_flood['total'] > 30].sort_values('total', ascending=False) \
        .assign(country=lambda x: pd.Categorical(x['country'], categories=x['country'][::-1], ordered=True)) \
        .plot(kind='barh', x='country', y='total', color='skyblue', legend=False)
    plt.title("Total number of Floods per country\nfrom 1901 to 2021")
    plt.xlabel("Total")
    plt.ylabel("Country")
    plt.show()
else:
    print("No data available for Floods.")
```

```python
disasters_columns3 = pd.read_csv('disasters_columns_clean.csv')
disasters_columns3
```

```python
import geopandas as gpd
import pandas as pd


# Filter records from 2000 onwards
df_filtered = disasters_columns3[disasters_columns3['year'] >= 2021]

# Aggregate to find the most common disaster type per country
most_common_disasters = df_filtered.groupby('country')['disaster_type'].agg(lambda x: x.value_counts().idxmax())

# Load a GeoDataFrame containing country geometries
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Merge the most common disaster data with the world GeoDataFrame
world = world.merge(most_common_disasters, how="left", left_on="name", right_index=True)

# Plot the map
world.plot(column='disaster_type', legend=True)

fig, ax = plt.subplots(figsize=(15, 10))

# Plotar o mapa with a bigher size
world.plot(column='disaster_type', legend=True, ax=ax)

# Show the map
plt.show()
```
