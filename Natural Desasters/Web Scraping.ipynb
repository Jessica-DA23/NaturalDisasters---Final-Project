## Comparison Between Countries - Web Scraping

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
from prophet import Prophet
```

```python
#!pip install beautifulsoup4 --upgrade

from bs4 import BeautifulSoup
import requests
import bs4

#China and French Polynesia
response = requests.get("https://www.dadosmundiais.com/comparacao-climatica.php?r1=china&r2=polinesia-francesa")

# Converta a resposta em um objeto BeautifulSoup
soup = bs4.BeautifulSoup(response.content, "html.parser")

# Encontre a tabela de dados climáticos
table = soup.find_all("table")[0]

# Extract the data from the table
data = []
for row in table.find_all("tr"):
    data.append([cell.text for cell in row.find_all("td")])

# Imprima os dados
print(data)
```

```python
# Remove empty lists and '\xa0' from the data
data_cleaned = [item for item in data if item and item[0] != '\xa0']

# Convert the list of lists into a DataFrame
china_vs_frenchp = pd.DataFrame(data_cleaned, columns=['Indicator', 'China', 'French Polynesia'])

# Print the DataFrame
china_vs_frenchp
```

```python
# Extract India and French Polynesia data
response = requests.get("https://www.dadosmundiais.com/comparacao-climatica.php?r1=india&r2=polinesia-francesa")

# Convert the respons using BeautifulSoup
soup = bs4.BeautifulSoup(response.content, "html.parser")

# Finding the table with the climate data
table = soup.find_all("table")[0]

# Extract the data from the table
data2 = []
for row in table.find_all("tr"):
    data2.append([cell.text for cell in row.find_all("td")])

# Imprima os dados
print(data2)
```

```python
# Remove empty lists and '\xa0' from the data
data_cleaned2 = [item for item in data2 if item and item[0] != '\xa0']

# Convert the list of lists into a DataFrame
india_vs_frenchp = pd.DataFrame(data_cleaned2, columns=['Indicator', 'India', 'French Polynesia'])

# Print the DataFrame
india_vs_frenchp
```
